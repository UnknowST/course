```python
from numpy import *  
from sklearn.datasets import load_iris
from sklearn import tree
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

iris = load_iris()
x_train, x_test,  y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2, random_state = 7)
#clf = tree.DecisionTreeClassifier()   # 此时使用的分类依据是 基尼指数
clf = tree.DecisionTreeClassifier( criterion='entropy')  #信息熵
clf = clf.fit(x_train, y_train)
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True) 
graph = graphviz.Source(dot_data) 
graph.render("预剪枝前")

depth = clf.get_depth()
leaf_node = clf.apply(iris.data)
#-----观察各个叶子节点上的样本个数---------
df  = pd.DataFrame({"leaf_node":leaf_node,"num":np.ones(len(leaf_node)).astype(int)})
df  = df.groupby(["leaf_node"]).sum().reset_index(drop=False)
df  = df.sort_values(by='num').reset_index(drop=True)
print("\n==== 树深度：",depth," ============")
print("==各个叶子节点上的样本个数：==")
print(df)

ans=clf.score(x_test,y_test)
print('Accuracy: %.3f ' % (ans.mean()))

#  预剪枝 用新调整的参数 训练模型
clf=tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=4,max_depth=3)
clf.fit(iris.data,iris.target)


dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True) 
graph = graphviz.Source(dot_data) 
graph.render("预剪枝后")

depth = clf.get_depth()
leaf_node = clf.apply(iris.data)
#-----观察各个叶子节点上的样本个数---------
df  = pd.DataFrame({"leaf_node":leaf_node,"num":np.ones(len(leaf_node)).astype(int)})
df  = df.groupby(["leaf_node"]).sum().reset_index(drop=False)
df  = df.sort_values(by='num').reset_index(drop=True)
print("\n==== 树深度：",depth," ============")
print("==各个叶子节点上的样本个数：==")
print(df)

ans=clf.score(x_test,y_test)
print('Accuracy: %.3f ' % (ans.mean()))
```

```python
# 后剪枝
 # -*- coding: utf-8 -*-
from sklearn.datasets import load_iris

from sklearn import tree
import numpy as np
 
#--------数据准备-----------------------------------
iris = load_iris()                          # 加载数据



x_train, x_test,  y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2, random_state = 7)
#-------模型训练---------------------------------
clf = tree.DecisionTreeClassifier(criterion='gini', min_samples_split=3,random_state=0,ccp_alpha=0)        
clf = clf.fit(x_train, y_train)     
#-------计算ccp路径------------------------------
pruning_path = clf.cost_complexity_pruning_path(x_train, y_train)
 
#-------打印结果---------------------------------   
print("\n====CCP路径=================")
print("ccp_alphas:",pruning_path['ccp_alphas'])
print("impurities:",pruning_path['impurities'])    
 
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True) 
graph = graphviz.Source(dot_data) 
graph.render("iris后剪枝前")

ans=clf.score(x_test,y_test)
print('Accuracy: %.3f ' % (ans.mean()))



#------设置alpha对树后剪枝-----------------------
clf = tree.DecisionTreeClassifier(criterion='gini', min_samples_split=3,random_state=0,ccp_alpha=0.015)        
clf = clf.fit(x_train, y_train) 
#------自行计算树纯度以验证-----------------------
is_leaf =clf.tree_.children_left ==-1
tree_impurities = (clf.tree_.impurity[is_leaf]* clf.tree_.n_node_samples[is_leaf]/len(y_train)).sum()
#-------打印结果--------------------------- 
print("\n==设置alpha=0.015剪枝后的树纯度：=========\n",tree_impurities)

import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True) 
graph = graphviz.Source(dot_data) 
graph.render("iris后剪枝后")

ans=clf.score(x_test,y_test)
print('Accuracy: %.3f ' % (ans.mean()))
```
